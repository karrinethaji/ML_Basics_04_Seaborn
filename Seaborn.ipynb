## Seaborn:

Seaborn is a Python data visualization library based on matplotlib. It provides a high-level interface for drawing attractive and informative statistical graphics.

```python
# Lets import seaborn package

import seaborn as sns
```

```python
# we have multiple datasets available under seaborn library, Let's import "iris" dataset

sns.load_dataset("iris")
```

```python
# we can plot the scatter plot using seaborn library as well
# To find the relation between two numerical features then we can plot the scatter plot and then we can understand the 
# correlatrion between two features

sns.scatterplot(x=iris.sepal_length,y=iris.petal_length)

# sns.scatterplot(x=iris['sepal_length'],y=iris['petal_length'])

# By seeing the below picture we can understand as sepal_length increases the petal_length also increasing that means these 
# two features having the positive correlation.
```

```python
print(sns.scatterplot(x=iris['sepal_length'],y=iris['petal_length']))
```

```python
print(sns.scatterplot(y=iris['sepal_length'],x=iris['petal_length']))
```

```python
# Lineplot graph

# Draw a line plot with possibility of several semantic groupings.

sns.lineplot(x=iris['sepal_length'],y=iris['petal_length'])
```

```python
# Join plot

# Draw a plot of two variables with bivariate and univariate graphs.

# This function provides a convenient interface to the :class:`JointGrid`
# class, with several canned plot kinds. This is intended to be a fairly
# lightweight wrapper; if you need more flexibility, you should use
# :class:`JointGrid` directly.

sns.jointplot(iris)
```

```python
# Joint plot:

# A join plot allows to study the relationship between two numerical features. The control chart display their correlation. It is
# Usually a scatter plot, a hexbin plot, a 2D histogram or 2d density plot

sns.jointplot(x=iris.sepal_length,y=iris.petal_length)
```

```python
#count plot : It is used to count the number of records available for each numerical feature

# Show the counts of observations in each categorical bin using bars.

# A count plot can be thought of as a histogram across a categorical, instead
# of quantitative, variable. The basic API and options are identical to those
# for :func:`barplot`, so you can compare counts across nested variables.

sns.countplot(iris)
```

```python
sns.countplot(x="species",data=iris) # No need to specify the "y" value because the "y" represents the count always
```

```python
# histogram

# Plot univariate or bivariate histograms to show distributions of datasets.

# A histogram is a classic visualization tool that represents the distribution
# of one or more variables by counting the number of observations that fall within
# discrete bins.

sns.histplot(iris.species) # histogram will provide the count of records for specified feature

# we have three types of species available under the iris.species column, the count of each species is 50
```

```python
sns.histplot(iris.sepal_length,kde=True)
```

```python
# distplot : It will be deprecated from the seaborn package

# It is used to check the distribution of feature

sns.distplot(iris.sepal_length)
```

```python
#displot --> It works same as histogram (histplot)

# Figure-level interface for drawing distribution plots onto a FacetGrid.

# This function provides access to several approaches for visualizing the
# univariate or bivariate distribution of data, including subsets of data
# defined by semantic mapping and faceting across multiple subplots. The
# ``kind`` parameter selects the approach to use:

sns.displot(iris.sepal_length)
```

```python
# violin plot

# Draw a combination of boxplot and kernel density estimate.

# A violin plot plays a similar role as a box and whisker plot. It shows the
# distribution of quantitative data across several levels of one (or more)
# categorical variables such that those distributions can be compared. Unlike
# a box plot, in which all of the plot components correspond to actual
# datapoints, the violin plot features a kernel density estimation of the
# underlying distribution.

sns.violinplot(data=tips)
```

```python
sns.violinplot(x="total_bill",y="day",data=tips)
```

```python
sns.violinplot(x="total_bill",y="day",hue="smoker",data=tips,palette="rainbow")
```

```python
tips
```

```python
# Bar Plot

# We can plot the bar graoh by providing the both x and y variables. Atleast one feature must be numeric.

# It is important to keep in mind that a bar plot shows only the mean ( or other estimator) value, but in many cases it may be
# more informatove to show the distribution of value at each value of the categorical variables. In that case, other approaches
# such as violin plot or box plot may be more appropriate.

sns.barplot(x="total_bill",y=tips.sex,data=tips)

print("The sum of toatl bill for male:",sum(tips[tips['sex']=="Male"]["total_bill"]))

print("Count of male:",tips.sex.value_counts()["Male"]) # or tips.sex.value_counts()[0]

print("Average of total_bill for men:",(sum(tips[tips['sex']=="Male"]["total_bill"])/(tips.sex.value_counts()["Male"])))

print("The sum of toatl bill for female:",sum(tips[tips['sex']=="Female"]["total_bill"]))

print("Count of Female:",tips.sex.value_counts()["Female"]) # or tips.sex.value_counts()[1]

print("Average of total_bill for Female:",(sum(tips[tips['sex']=="Female"]["total_bill"])/(tips.sex.value_counts()["Female"])))

# The above average represented in the below graphs
```

```python

sns.barplot(y="total_bill",x=tips.smoker,data=tips)
```

```python
# Plotting subplots using seaborn

import matplotlib.pyplot as plt

fig,axes=plt.subplots(2,1,figsize=(10,20))

sns.barplot(y="total_bill",x=tips.smoker,data=tips,ax=axes[0])
sns.violinplot(x="total_bill",y="day",hue="smoker",data=tips,palette="rainbow",ax=axes[1])
axes[0].set_title("This is bar plot")
axes[1].set_title("This is violin plot")
```
